```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

1- savol Berilgan datasetdan 2 ta ixtiroiy ustunni tanlab olib (10,9) o'lchamda grafik ko'rinishini hosil qilib uni tahlil qiling

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df
```

```python
#1-savolga javob
plt.figure(figsize=(10,9))
sns.scatterplot(data=df,x=df['year'],y=df['mileage'],hue="fuelType")
plt.show()
```
