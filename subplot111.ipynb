```python
import numpy as np
import pandas as ps
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
plt.figure(figsize=(12,6))
# sin(x) grafigi

plt.plot(x, sin, linewidth=1.5,
                linestyle='--',
                color='blue',
                marker='h')
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
plt.figure(figsize=(12,6))
# sin(x) grafigi

plt.plot(x, sin, linewidth=1.5,
                linestyle='--',
                color='blue',
                marker='h')
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', linestyle='--')
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
plt.figure(figsize=(12,6))
# sin(x) grafigi

plt.plot(x, sin, linewidth=1.5,
                linestyle='--',
                color='blue',
                marker='h')
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', linestyle='--')
# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         marker='v')
plt.title('cos(x) funksiya grafigi')
plt.grid(color='tomato', linestyle='--')
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)
plt.figure(figsize=(12,6))
# sin(x) grafigi
plt.subplot(2,1,1)
plt.plot(x, sin, linewidth=1.5,
                linestyle='--',
                color='blue',
                marker='h')
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', linestyle='--')
# cos(x) grafigi
plt.subplot(2,1,2)
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         marker='v')
plt.title('cos(x) funksiya grafigi')
plt.grid(color='tomato', linestyle='--')
plt.show()
```
